Aggregation Feature - Last 3,6,9,12

```python
#Aggregation features for the last [3,6,9,12]
# Set the reference date as April 2018
reference_date = pd.Timestamp("2018-03-31")
 
# Function to calculate stats for last N months
def last_n_months_stats(df, months, num_columns):
    start_date = reference_date - pd.DateOffset(months=months)
   
    # Filter only data within the required time range
    df_filtered = df[(df["S_2"] > start_date) & (df["S_2"] <= reference_date)]
 
    # Aggregate per customer
    agg_df = df_filtered.groupby("customer_ID")[num_columns].agg(["mean", "min", "max", "std"])
 
    # Flatten column names (remove sub-columns)
    agg_df.columns = [f"{col}_{stat}_{months}m" for col, stat in agg_df.columns]
 
    return agg_df.reset_index()
 
# Compute stats for 3, 6, 9, and 12 months
stats_3m = last_n_months_stats(df_encoded, 3, num)
stats_6m = last_n_months_stats(df_encoded, 6, num)
stats_9m = last_n_months_stats(df_encoded, 9, num)
stats_12m = last_n_months_stats(df_encoded, 12, num)
 
# Merge all results into one final dataset (1 row per customer)
df_final = stats_3m.merge(stats_6m, on="customer_ID", how="left")\
                    .merge(stats_9m, on="customer_ID", how="left")\
                    .merge(stats_12m, on="customer_ID", how="left")
 
print(df_final.shape)  # Should be (number of unique customers, aggregated features)
print(df_final.head())  # Check if column names are correct
```

Rate of Change Feature

```python
# Define function to calculate Rate of Change for each customer

# Define reference dates
reference_date_2018 = pd.Timestamp("2018-03-31")
reference_date_2017 = pd.Timestamp("2017-03-31")

# Function to get latest available data before or on a given date
def get_latest_available(group, ref_date, num_columns):
    # Filter records on or before the reference date and get the latest one
    latest_record = group[group["S_2"] <= ref_date].sort_values("S_2").iloc[-1:]  # Latest row
    if latest_record.empty:
        return pd.Series(np.nan, index=num_columns)  # Return NaNs if no data
    return latest_record[num_columns].iloc[0]  # Get values for the specified columns

# Function to calculate Rate of Change (ROC)
def calculate_rate_of_change(group, num_columns):
    # Get the latest available values for April 2018 and April 2017
    apr_2018_values = get_latest_available(group, reference_date_2018, num_columns)
    apr_2017_values = get_latest_available(group, reference_date_2017, num_columns)
    
    # Calculate the rate of change: (Apr 2018 value - Apr 2017 value) / Apr 2017 value
    rate_of_change = (apr_2018_values - apr_2017_values) / apr_2017_values
    
    # Rename columns to include the rate of change calculation
    rate_of_change.index = [f'{col}_rate_of_change' for col in rate_of_change.index]

    return rate_of_change

# Apply the function to each customer
roc = df_encoded.groupby("customer_ID").apply(lambda group: calculate_rate_of_change(group, num))

# Reset the index to have a clean DataFrame
roc = roc.reset_index()

# Show results
print(roc.head())  # Display the first few rows of the result


```

Ratio Feature

```python
# Identify spend (S*), balance (B*), and payment (P*) columns from `num`
spend_cols = [col for col in num if col.startswith('S')]
balance_cols = [col for col in num if col.startswith('B')]
payment_cols = [col for col in num if col.startswith('P')]

# Function to calculate Spend-to-Balance and Payment-to-Spend ratios
def calculate_ratios(group):
    # 4️⃣ Spend-to-Balance Ratio (sum of all S* columns / sum of all B* columns)
    total_spend = group[spend_cols].sum().sum() if spend_cols else np.nan
    total_balance = group[balance_cols].sum().sum() if balance_cols else np.nan
    spend_to_balance = total_spend / total_balance if total_balance and total_balance != 0 else np.nan

    # 5️⃣ Payment-to-Spend Ratio (sum of all P* columns / sum of all S* columns)
    total_payment = group[payment_cols].sum().sum() if payment_cols else np.nan
    payment_to_spend = total_payment / total_spend if total_spend and total_spend != 0 else np.nan

    return pd.Series({
        'spend_to_balance_ratio': spend_to_balance,
        'payment_to_spend_ratio': payment_to_spend
    })

# Apply the function to each customer
ratios = df_encoded.groupby('customer_ID').apply(calculate_ratios)

# Show results
print(ratios)
```

Volatility Feature

```python


# Identify spend (S*), balance (B*), and payment (P*) columns from `num`
spend_cols = [col for col in num if col.startswith('S') and pd.api.types.is_numeric_dtype(df_encoded[col])]
balance_cols = [col for col in num if col.startswith('B') and pd.api.types.is_numeric_dtype(df_encoded[col])]
payment_cols = [col for col in num if col.startswith('P') and pd.api.types.is_numeric_dtype(df_encoded[col])]

# Group by customer_ID and calculate volatility features for each customer
def calculate_volatility(group):
    # For spend: aggregate mean and std across all spend columns
    spend_mean = group[spend_cols].mean(axis=1).mean()  # mean of each row and then mean of all rows
    spend_std = group[spend_cols].std(axis=1).mean()    # std of each row and then mean of all rows

    # For balance: aggregate mean and std across all balance columns
    balance_mean = group[balance_cols].mean(axis=1).mean()
    balance_std = group[balance_cols].std(axis=1).mean()

    # For payment: aggregate mean and std across all payment columns
    payment_mean = group[payment_cols].mean(axis=1).mean()
    payment_std = group[payment_cols].std(axis=1).mean()

    # Calculate volatility (Coefficient of Variation)
    spend_volatility = spend_std / spend_mean if spend_mean != 0 else np.nan
    balance_volatility = balance_std / balance_mean if balance_mean != 0 else np.nan
    payment_volatility = payment_std / payment_mean if payment_mean != 0 else np.nan

    return pd.Series({
        'spend_volatility': spend_volatility,
        'balance_volatility': balance_volatility,
        'payment_volatility': payment_volatility
    })

# Group by customer_ID and apply the volatility calculation
volatility_features = df_encoded.groupby('customer_ID')[num].apply(calculate_volatility)

# Show the result with customer_ID as index
print(volatility_features.head())
```

**Categorical Features**

```python
# Set the reference date as April 30, 2018
reference_date = pd.Timestamp('2018-03-31')

def last_n_months_categorical_stats(df, months, cat_columns):
    # Calculate the start date for the given number of months
    start_date = reference_date - pd.DateOffset(months=months)
   
    # Filter only data within the required date range
    df_filtered = df[(df["S_2"] > start_date) & (df["S_2"] <= reference_date)]
   
    # Compute response rate (percentage of times the value is 1)
    response_rate = df_filtered.groupby("customer_ID")[cat_columns].sum() / df_filtered.groupby("customer_ID")[cat_columns].count()
   
    # Compute ever response (whether 1 appeared at least once)
    ever_response = df_filtered.groupby("customer_ID")[cat_columns].apply(lambda x: (x.sum() > 0).astype(int))
   
    # Rename columns
    response_rate.columns = [f"{col}_Response_Rate_{months}m" for col in response_rate.columns]
    ever_response.columns = [f"{col}_Ever_Response_{months}m" for col in ever_response.columns]
   
    # Merge response rate and ever response
    agg_df = response_rate.merge(ever_response, on="customer_ID")
   
    return agg_df.reset_index()

# Compute stats for 3, 6, 9, and 12 months using encoded_cols
stats_3m = last_n_months_categorical_stats(df_encoded, 3, encoded_cols)
stats_6m = last_n_months_categorical_stats(df_encoded, 6, encoded_cols)
stats_9m = last_n_months_categorical_stats(df_encoded, 9, encoded_cols)
stats_12m = last_n_months_categorical_stats(df_encoded, 12, encoded_cols)

# Merge the results into one final dataset
df_response = stats_3m.merge(stats_6m, on="customer_ID", how="left")\
                    .merge(stats_9m, on="customer_ID", how="left")\
                    .merge(stats_12m, on="customer_ID", how="left")

# Show the result
print(df_response.head())
```

```python
# Define the parameter grid
param_grid_xgb = {
    'n_estimators': [50, 100, 300],  # Number of trees
    'learning_rate': [0.01, 0.1],  # Learning rate
    'subsample': [0.5, 0.8],  # Percentage of observations used in each tree
    'colsample_bytree': [0.5, 1.0],  # Percentage of features used in each tree
    'scale_pos_weight': [1, 5, 10]  # Weight of default observations
}
```

```python
# Load previous results if available (to avoid re-running completed iterations)
csv_filename = "grid_search_results_xgb.csv"
try:
    results_df = pd.read_csv(csv_filename)
    completed_combinations = set(tuple(row) for row in results_df.iloc[:, :5].values)
except FileNotFoundError:
    results_df = pd.DataFrame(columns=[
        "Trees", "LR", "Subsample", "% Features", "Weight of Default",
        "AUC Train", "AUC Test 1", "AUC Test 2"
    ])
    completed_combinations = set()
```

```python
# Iterate through all combinations of parameters
for n_estimators, learning_rate, subsample, colsample_bytree, scale_pos_weight in itertools.product(
    param_grid_xgb['n_estimators'], param_grid_xgb['learning_rate'],
    param_grid_xgb['subsample'], param_grid_xgb['colsample_bytree'],
    param_grid_xgb['scale_pos_weight']
):
    # Check if this combination was already completed
    param_tuple = (n_estimators, learning_rate, subsample, colsample_bytree, scale_pos_weight)
    if param_tuple in completed_combinations:
        continue

    # Define model
    xgb_model = XGBClassifier(
        n_estimators=n_estimators,
        learning_rate=learning_rate,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        scale_pos_weight=scale_pos_weight,
        objective='binary:logistic',
        use_label_encoder=False,
        eval_metric='auc',
        random_state=42
    )

    # Train model
    xgb_model.fit(X_train_selected, y_train)

    # Get AUC scores
    auc_train = roc_auc_score(y_train, xgb_model.predict_proba(X_train_selected)[:, 1])
    auc_test1 = roc_auc_score(y_test1, xgb_model.predict_proba(X_test1_selected)[:, 1])
    auc_test2 = roc_auc_score(y_test2, xgb_model.predict_proba(X_test2_selected)[:, 1])

    # Append results to dataframe
    results_df.loc[len(results_df)] = [
        n_estimators, learning_rate, subsample, colsample_bytree, scale_pos_weight,
        auc_train, auc_test1, auc_test2
    ]

    # Save after each iteration to avoid losing progress
    results_df.to_csv(csv_filename, index=False)

    print(f"Completed: {param_tuple}, AUC Test1: {auc_test1:.4f}, AUC Test2: {auc_test2:.4f}")

print("\n Grid Search Completed! Check results in", csv_filename)
```

```python
# Step 1: Handle missing values and replace them with 0 (SimpleImputer)
imputer = SimpleImputer(strategy='constant', fill_value=0)
X_train_imputed = imputer.fit_transform(X_train_selected)
X_test1_imputed = imputer.transform(X_test1_selected)
X_test2_imputed = imputer.transform(X_test2_selected)
```

```python
# Step 2: Outlier treatment (Capping and Flooring at the 1st and 99th percentiles)
def cap_and_floor(X):
    lower_percentile = np.percentile(X, 1, axis=0)
    upper_percentile = np.percentile(X, 99, axis=0)
    X_clipped = np.clip(X, lower_percentile, upper_percentile)
    return X_clipped

X_train_clipped = cap_and_floor(X_train_imputed)
X_test1_clipped = cap_and_floor(X_test1_imputed)
X_test2_clipped = cap_and_floor(X_test2_imputed)
```

```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics import roc_auc_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import backend as K

# === Grid Search Parameters ===
nn_hidden_layers_list = [2, 4]
nn_nodes_list = [4, 6]
nn_activation_list = ['relu', 'tanh']
nn_dropout_list = [0.5, 0.0]  # 50%, 100% (no dropout)
nn_batch_size_list = [100, 10000]
nn_epochs = 20
nn_results_path = 'nn_grid_results.csv'
nn_detailed_log_path = 'nn_grid_detailed_log.csv'

# === Load Previous Results if Exist ===
if os.path.exists(nn_results_path):
    result_df_nn = pd.read_csv(nn_results_path)
else:
    result_df_nn = pd.DataFrame(columns=[
        'HL', '# Node', 'Activation Function', 'Dropout', 'Batch Size',
        'AUC Train', 'AUC Test1', 'AUC Test2'
    ])

# === Create Detailed Log File If Not Exists ===
if not os.path.exists(nn_detailed_log_path):
    pd.DataFrame(columns=[
        'HL', '# Node', 'Activation Function', 'Dropout', 'Batch Size',
        'AUC Train', 'AUC Test1', 'AUC Test2'
    ]).to_csv(nn_detailed_log_path, index=False)    



# === Build Neural Network Model ===
def build_model_nn(n_hidden, n_nodes, activation, dropout, input_dim):
    model_nn = Sequential()
    model_nn.add(Dense(n_nodes, activation=activation, input_dim=input_dim))
    if dropout > 0:
        model_nn.add(Dropout(dropout))

    for _ in range(n_hidden - 1):
        model_nn.add(Dense(n_nodes, activation=activation))
        if dropout > 0:
            model_nn.add(Dropout(dropout))

    model_nn.add(Dense(1, activation='sigmoid'))
    model_nn.compile(optimizer=Adam(), loss=BinaryCrossentropy(), metrics=[])
    return model_nn

# === Begin Grid Search ===
for nn_hl in nn_hidden_layers_list:
    for nn_node in nn_nodes_list:
        for nn_act in nn_activation_list:
            for nn_drop in nn_dropout_list:
                for nn_bs in nn_batch_size_list:

                    # Skip if already computed
                    if ((result_df_nn['HL'] == nn_hl) & 
                        (result_df_nn['# Node'] == nn_node) &
                        (result_df_nn['Activation Function'] == nn_act) &
                        (result_df_nn['Dropout'] == f"{int(nn_drop*100)}%") &
                        (result_df_nn['Batch Size'] == nn_bs)).any():
                        continue

                    print(f"Training NN | HL:{nn_hl}, Nodes:{nn_node}, Act:{nn_act}, Dropout:{nn_drop}, Batch:{nn_bs}")

                    # Build and train model
                    model_nn = build_model_nn(nn_hl, nn_node, nn_act, nn_drop, X_train_scaled.shape[1])
                    early_stop_nn = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)

                    model_nn.fit(
                        X_train_scaled, y_train,
                        batch_size=nn_bs,
                        epochs=nn_epochs,
                        verbose=0,
                        callbacks=[early_stop_nn]
                    )

                    # Evaluate
                    try:
                        auc_train_nn = roc_auc_score(y_train, model_nn.predict(X_train_scaled))
                        auc_test1_nn = roc_auc_score(y_test1, model_nn.predict(X_test1_scaled))
                        auc_test2_nn = roc_auc_score(y_test2, model_nn.predict(X_test2_scaled))
                    except Exception as e:
                        print(f"Error in AUC calculation: {e}")
                        auc_train_nn, auc_test1_nn, auc_test2_nn = np.nan, np.nan, np.nan

                    new_row_nn = pd.DataFrame([{
                    'HL': nn_hl,
                    '# Node': nn_node,
                    'Activation Function': nn_act,
                    'Dropout': f"{int(nn_drop*100)}%",
                    'Batch Size': nn_bs,
                    'AUC Train': auc_train_nn,
                    'AUC Test1': auc_test1_nn,
                    'AUC Test2': auc_test2_nn
                }])

                # === Append to Summary & Detailed Logs ===
                    result_df_nn = pd.concat([result_df_nn, new_row_nn], ignore_index=True)
                    new_row_nn.to_csv(nn_detailed_log_path, mode='a', header=False, index=False)

# === Save Final Summary ===
result_df_nn.to_csv(nn_results_path, index=False)
print("NN Grid Search Complete. Results saved to:")
print(f"Summary: {nn_results_path}")
print(f"Detailed log: {nn_detailed_log_path}")
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import save_model

# === Load Grid Search Results for NN ===
nn_results_df = pd.read_csv("nn_grid_results.csv")

# === Compute Average AUC and Variance for Bias-Variance Trade-off ===
nn_results_df["Avg AUC"] = nn_results_df[["AUC Train", "AUC Test1", "AUC Test2"]].mean(axis=1)
nn_results_df["AUC Variance"] = nn_results_df[["AUC Train", "AUC Test1", "AUC Test2"]].var(axis=1)

# === Identify Top 5 Based on Avg AUC, Then Pick Model with Lowest Variance ===
top_5_models = nn_results_df.nlargest(5, "Avg AUC")
best_model_row = nn_results_df.loc[top_5_models["AUC Variance"].idxmin()]

print("Best Neural Network Hyperparameters Chosen:")
print(best_model_row)
```

```python

# === Extract Hyperparameters ===
num_layers = int(best_model_row["HL"])
num_nodes = int(best_model_row["# Node"])
activation_fn = best_model_row["Activation Function"].lower()
dropout_rate = 0.0 if best_model_row["Dropout"] == "100%" else float(best_model_row["Dropout"].strip('%')) / 100
batch_size = int(best_model_row["Batch Size"])

# === Build Neural Network ===
def build_nn_model(input_dim, hidden_layers, nodes, activation, dropout):
    model = Sequential()
    model.add(Dense(nodes, activation=activation, input_dim=input_dim))
    if dropout > 0:
        model.add(Dropout(dropout))
    for _ in range(hidden_layers - 1):
        model.add(Dense(nodes, activation=activation))
        if dropout > 0:
            model.add(Dropout(dropout))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=Adam(), loss=BinaryCrossentropy())
    return model

# === Train Final Model ===
final_model = build_nn_model(
    input_dim=X_train_scaled.shape[1],
    hidden_layers=num_layers,
    nodes=num_nodes,
    activation=activation_fn,
    dropout=dropout_rate
)

early_stop = EarlyStopping(monitor="loss", patience=3, restore_best_weights=True)

final_model.fit(
    X_train_scaled, y_train,
    epochs=20,
    batch_size=batch_size,
    verbose=1,
    callbacks=[early_stop]
)

# === Save Final Trained Model ===
final_model.save("final_nn_model.h5")
print("Final model saved as 'final_nn_model.h5'")
```
